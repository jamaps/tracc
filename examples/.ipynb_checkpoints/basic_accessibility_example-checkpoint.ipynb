# Measuring Transport Accessbility

A simple eexample using `tracc` for the Boston region

```python
import sys
sys.path.append("..")

import tracc
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib

%matplotlib inline
```

Load in the destination data. For this example, these are job counts by block group from the the LEHD and SNAP grocery stores.

```python
dfo = pd.merge(
        pd.read_csv("test_data/boston/destination_employment_lehd.csv"),
        pd.read_csv("test_data/boston/destination_groceries_snap.csv"),
        how = "outer",
        left_on = "block_group_id",
        right_on = "GEOID"
)
```

Create a supply object, for only the columns in which we want to compute accessibility for. In this case, all jobs from the LEHD (var C000), and SNAP grocery stores.

```python
dfo = tracc.supply(
    supply_df = dfo,
    columns = ["block_group_id","C000","snap"]
    )
dfo.data.head()
```

Now let's load in travel time data. Here we load in transit travel times that were previously computed in OpenTripPlanner (capped at 120 minutes) for June 30, 2020.

```python
dft = tracc.costs(
    pd.read_csv("test_data/boston/transit_time_matrix_8am_30_06_2020.zip", compression='zip')
    )
dft.data.time = dft.data.time / 60 # converting time from seconds to minutes
dft.data.time = dft.data.time.round(1) # rounding to just one decimal place
dft.data.head()
```

Computing impedence function based on a 45 minute travel time threshold 

```python
dft.impedence_calc(
    cost_column = "time",
    impedence_func = "cumulative",
    impedence_func_params = 45,
    output_col_name = "fCij_c45",
    prune_output = False
)
dft.data.head()
```

Setting up the accessibility object. This includes joining the destination data to the travel time data

```python
acc = tracc.accessibility(
        travelcosts_df = dft.data,
        supply_df = dfo.data,
        travelcosts_ids = ["o_block","d_block"],
        supply_ids = "block_group_id"
    )
```

Measuring potential accesibility to jobs, using a 45 minute cumulative impedence function.

```python
dfa = acc.potential(
        opportunity = "C000",
        impedence = "fCij_c45"
        )
dfa.head()
```

Let's try to visualize the result, first by loadidng in shape for the Block Groups, joining our output to block groups, then plotting

```python
bg = gpd.read_file("test_data/boston/block_group_poly.geojson")
bg = bg.merge(dfa, left_on='GEOID', right_on = "o_block", how = "left")
```

```python
bg.plot(column='A_C000_fCij_c45', figsize=(8, 8), scheme='quantiles', legend=True, legend_kwds={'loc': 'upper left'});
```

Now let's try to measure access to grocery stores, basesd on minumum travel time 
(e.g. what is the minimum travel time to reach 2 grocery stores)

```python
dfm = acc.mintravelcost(
        travelcost = "time",
        opportunity = "snap",
        min_n = 2
        )
dfm.head()
```

```python
bg = gpd.read_file("test_data/boston/block_group_poly.geojson")
bg = bg.merge(dfm, left_on='GEOID', right_on = "o_block", how = "left")
```

```python
bg.plot(column='A_mintravelcost_time_snap_2', figsize=(8, 8), scheme='NaturalBreaks', legend=True, legend_kwds={'loc': 'upper left'};
```
